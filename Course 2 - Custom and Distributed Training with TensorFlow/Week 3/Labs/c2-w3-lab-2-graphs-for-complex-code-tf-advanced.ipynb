# Autograph: Graphs for complex code

## Imports

```python
import tensorflow as tf
```

```python
a = tf.Variable(1.0)
b = tf.Variable(2.0)

@tf.function
def f(x,y):
    a.assign(y * b)
    b.assign_add(x * a)
    return a + b

print(f(1.0, 2.0))

print(tf.autograph.to_code(f.python_function))
```

```python
@tf.function
def sign(x):
  if x > 0:
    return 'Positive'
  else:
    return 'Negative or zero'

print("Sign = {}".format(sign(tf.constant(2))))
print("Sign = {}".format(sign(tf.constant(-2))))

print(tf.autograph.to_code(sign.python_function))
```

```python
@tf.function
def f(x):
  while tf.reduce_sum(x) > 1:
    tf.print(x)
    x = tf.tanh(x)
  return x

#f(tf.random.uniform([5]))

print(tf.autograph.to_code(f.python_function))
```

```python
@tf.function
def sum_even(items):
  s = 0
  for c in items:
    if c % 2 > 0:
      continue
    s += c
  return s

print(tf.autograph.to_code(sum_even.python_function))
```

## Print statements

```python
def f(x):
  print("Traced with", x)

for i in range(5):
  f(2)
f(3)
```

```python
@tf.function
def f(x):
  print("Traced with", x)

for i in range(5):
  f(2)
f(3)
```

```python
@tf.function
def f(x):
    print("Traced with", x)
    # added tf.print
    tf.print("Executed with", x)
    print("(print) Executed with", x)


for i in range(5):
  f(2)
f(3)
```

## Avoid defining variables inside the function

```python
def f(x):
    v = tf.Variable(1.0)
    v.assign_add(x)
    return v

print(f(1))
```

```python
@tf.function
def f(x):
  v = tf.Variable(1.0)
  v.assign_add(x)
  return v
print(f(1))
```

```python
# define the variables outside of the decorated function
v = tf.Variable(1.0)

@tf.function
def f(x):
  return v.assign_add(x)

print(f(5))
```
